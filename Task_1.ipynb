```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
# total revenue per product
total_revenue = df.groupby("Product")["Revenue"].sum()
print("Total Revenue per product:\n",total_revenue)

avg_units_sold = df.groupby("Product")["Units_Sold"].mean()
print("\nAverage units sold per product:\n",avg_units_sold)
```

```python
for product in df["Product"].unique():
    product_data = df[df["Product"]==product]
    plt.plot(product_data["Month"],product_data["Revenue"],marker='o',label=product)
plt.title("Monthly Revenue Trend (Jan-Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

```python
total_revenue.plot(kind="bar",color=["skyblue","orange","green"])
plt.title("Total Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
total_revenue.plot(kind="pie",autopct="%1.1f%%",startangle=90)
plt.title("Revenue contribution by Product")
plt.ylabel("")
plt.show()
```
