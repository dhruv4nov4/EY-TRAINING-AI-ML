```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Rahul", "Priya", "Arjun", "Neha", "Vikram"],
    "Age": [21, 22, 20, 23, 21],
    "Course": ["AI", "ML", "Data Science", "AI", "ML"],
    "Marks": [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)
print(df)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
avg_marks= df.groupby("Course")["Marks"].mean()
avg_marks.plot(kind="bar",color="skyblue",edgecolor="red")
plt.title("avg marks by course")
plt.xlabel("course")
plt.ylabel("avg marks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt
 
# Sample sales dataset

data = {

    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],

    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],

    "Units_Sold": [120, 135, 150, 200, 220, 210],

    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]

}
 
df = pd.DataFrame(data)
df
avg_marks= df.groupby("Product")["Revenue"].mean()
avg_marks.plot(kind="bar",color="green",edgecolor="black")
plt.title("Sales")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
